# Needed libraries:

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Define the path:

```python
archive=r"path"
```

# Reading images, resizing and saving them:

```python
for img in os.listdir( archive + "\\original"):
    img_array = cv2.imread(archive + "\\original\\" + img)
    
    img_array = cv2.resize(img_array, (128,128))
    lr_img_array = cv2.resize(img_array,(32,32))
    cv2.imwrite(archive+ "\\high\\" + img, img_array)
    cv2.imwrite(archive+ "\\low\\"+ img, lr_img_array)
```
